```python
import pandas as pd
import numpy as np
```

# 말뭉치

```python
train=pd.read_excel('감성대화말뭉치(최종데이터)_Training.xlsx')
val=pd.read_excel('감성대화말뭉치(최종데이터)_Validation.xlsx')
```

```python
train = train.fillna("")
val = val.fillna("")
```

```python
train['사람문장'] = train['사람문장1']
val['사람문장'] = val['사람문장1']
```

```python
train = train[["감정_대분류","사람문장"]]
train.columns = ['emotion', 'sentence']

val = val[["감정_대분류","사람문장"]]
val.columns = ['emotion', 'sentence']
```

```python
train["emotion"].value_counts()
val["emotion"].value_counts()
```

# 단발성

```python
data_single = pd.read_excel('한국어_단발성_대화_데이터셋.xlsx')
```

```python
data_single = data_single[data_single["Emotion"]!="놀람"]
data_single = data_single[data_single["Emotion"]!="중립"]
data_single = data_single[data_single["Emotion"]!="혐오"]
```

```python
data_single["Emotion"] = np.where(data_single["Emotion"].str.match("공포"),"불안",data_single["Emotion"])
```

```python
data_single = data_single[["Emotion", "Sentence"]]
data_single.columns = ['emotion', 'sentence']
```

```python
data_single["emotion"].value_counts()
```

# 연속형

```python
data_multi = pd.read_excel('한국어_연속적_대화_데이터셋.xlsx')
```

```python
data_multi = data_multi[["Unnamed: 2", "Unnamed: 1"]]
data_multi.columns = ['emotion', 'sentence']
```

```python
data_multi = data_multi[data_multi["emotion"].isin(["중립","놀람","분노","슬픔","행복","혐오","공포"])]
```

```python
data_multi["emotion"].value_counts()
```

```python
data_multi = data_multi[data_multi["emotion"]!="놀람"]
data_multi = data_multi[data_multi["emotion"]!="중립"]
data_multi = data_multi[data_multi["emotion"]!="혐오"]
```

```python
data_multi["emotion"] = np.where(data_multi["emotion"].str.match("공포"),"불안",data_multi["emotion"])
```

```python
data_multi["emotion"].value_counts()
```

# 합치기

```python
data = pd.concat([train,val,data_single,data_multi])
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data["emotion"].value_counts()
```

# 모델링 전처리

```python
data['emotion'] = data['emotion'].replace({'분노': 1, '슬픔': 2,
                                           '불안': 3, '상처': 4,
                                           '당황': 5, '기쁨': 6,
                                           '행복': 7})
```

```python
data.to_csv("emotionDataFin.csv")
```
